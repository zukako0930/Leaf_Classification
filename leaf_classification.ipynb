```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
train
```

```python
del train['id']
train
```

```python
X_1 = train.iloc[:,1:193].values
X_1
```

```python
y_1 = train['species'].values
y_1
```

```python
# データの標準化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_1)
X_1_std = scaler.transform(X_1)
```

```python
# ホールドアウト法による分割
from sklearn.model_selection import train_test_split
X_1_train, X_1_test, y_1_train, y_1_test = train_test_split(X_1_std, y_1, test_size=0.3, random_state=0)
```

```python
# 線形SVCの学習
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_1_train, y_1_train)
```

```python
# 正答率を出力
print('train accuracy:{}'.format(svc.score(X_1_train,y_1_train)))
print('test accuracy:{}'.format(svc.score(X_1_test, y_1_test)))
```

```python
# いいスコアが出た
# 次に決定木で分類してみる
from sklearn.tree import DecisionTreeClassifier
tree_1 = DecisionTreeClassifier(random_state=0)
tree_1.fit(X_1_train, y_1_train)
```

```python
# 精度の確認
print('train accuracy:{}'.format(tree_1.score(X_1_train, y_1_train)))
print('test accuracy:{}'.format(tree_1.score(X_1_test, y_1_test)))
```

```python
# test精度が低い
# ランダムフォレストを試してみる（バギング）
from sklearn.ensemble import RandomForestClassifier
rfc_1 = RandomForestClassifier(random_state=0, n_estimators=150)
rfc_1.fit(X_1_train, y_1_train)
```

```python
# 精度の確認
print('train accuracy:{}'.format(rfc_1.score(X_1_train, y_1_train)))
print('test accuracy:{}'.format(rfc_1.score(X_1_test, y_1_test)))
```

```python
# 木の本数150本でよい精度が出た
# ロジスティック回帰による分類も試してみる
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1.0)
lr.fit(X_1_train, y_1_train)
```

```python
# 正答率の出力
print('train accuracy:{}'.format(lr.score(X_1_train, y_1_train)))
print('test accuracy:{}'.format(lr.score(X_1_test, y_1_test)))
```
